```python
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url = "https://www.tottus.com.pe/abarrotes-4010000/c/?sort=Recomendados"
```

```python
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
contenido = soup.find_all("li",{"class":"product"})
```

```python
for elem in contenido:
    producto = elem.find("div",{'class':"jsx-369422875 details"}).contents
    precio = elem.find("span",{"class":"list price medium cmrPrice"})
    tipoVenta = elem.find("div",{"class":"jsx-369422875 delivery-type"}).contents
    print("")
    if producto[0]:
        print(producto[0].text)
    else:
        print("None")
    if producto[1]:
        print(producto[1].text)
    else:
        print("None")
    if precio:
        print(precio.text)
    else:
        print("None")
    if tipoVenta[0]:
        print(tipoVenta[0]["class"][1])
    else:
        print("None")
    if tipoVenta[1]:
        print(tipoVenta[1]["class"][1])
    else:
        print("None")
    print("----")
```
