```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pythonjobs.github.io'
peticion = requests.get(url)
```

```python
soup = BeautifulSoup(peticion.content,"html.parser")
soup
```

```python
soup.find(id="content")
```

```python
result = soup.find_all("div",class_="job")
for elem in result:
    print(elem)
    print(end='\n'*2)
```

```python
for elem in result:
    title = elem.find('h1')
    location = elem.find_all('span')[0]
    fecha = elem.find_all('span')[1]
    print("Trabajo: "+title.contents[0].text)
    print(" Ubicacion: "+location.text)
    print(" FechaPublicaion: "+fecha.text)
    print(end='\n'*1)
    
```

```python
titulos = soup.find_all("h1",string= lambda texto : "scientist" in texto.lower())
```

```python
titulos[0].text
```

```python
job_list = soup.find_all('div',class_="job")
for elem in job_list:
    detail = elem.find("p",class_="detail")
    print(detail.text.strip())
    print(end="\n"*1)
```
