```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://remote.co/remote-jobs/developer/'
peticion = requests.get(url)
```

```python
soup = BeautifulSoup(peticion.content,'html.parser')
```

```python
soup
```

```python
card = soup.find_all("div",{"class":"card-body p-0"})[1]
```

```python
titulos = card.find_all('a')[1:]
```

```python
for elem in titulos:
    limp = elem.find('p',class_="m-0").contents[0]
    for i in limp:
        print(i.text)
```
